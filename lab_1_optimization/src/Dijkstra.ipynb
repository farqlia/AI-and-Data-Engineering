Czas zostanie znormalizowany tak, by wszystkie godziny były w przedziale [0:00; 23:59]

```python
def to_seconds(time: str) -> int:
    h, m, s = separate_time(time)
    return h * 3600 + m * 60 + s

def time_to_normalized_sec(time: str) -> int:
    return to_seconds(time) % (3600 * 24)
```

```python
connection_graph['departure_sec'] = connection_graph['departure_time'].apply(time_to_normalized_sec)
connection_graph['arrival_sec'] = connection_graph['arrival_time'].apply(time_to_normalized_sec)
```

```python
def diff(ts: Union[pd.Series, int], td: int) -> int:
    '''Function that returns difference between times expressed as seconds'''
    d = ts - td 
    if isinstance(ts, pd.Series):
        d[d < 0] += 24 * 3600
    elif d < 0:
        d += 24 * 3600
    return d 
```

```python
from typing import Union, List

class Graph:

    def __init__(self, conn_graph):
        self.conn_graph = conn_graph
        self.transform()

    def transform(self):
        self.conn_graph['departure_sec'] = self.conn_graph['departure_time'].apply(time_to_normalized_sec)
        self.conn_graph['arrival_sec'] = self.conn_graph['arrival_time'].apply(time_to_normalized_sec)

    def get_nodes(self):
        '''should we also add end_stop?'''
        return self.conn_graph['start_stop'].unique()
    
    def get_lines(self, start_stop, end_stop):
        '''Returns lines connecting two stops'''
        return self.conn_graph[(self.conn_graph['start_stop'] == start_stop) 
                               & (self.conn_graph['end_stop'] == end_stop)]['line'].unique()
    
    def get_stops(self, line):
        return self.conn_graph[self.conn_graph['line'] == line]['start_stop'].unique()

    def get_neighbours(self, start_stop: str) -> np.array:
        '''Returns neighbouring end stops'''
        # return self.conn_graph[self.conn_graph['start_stop'] == start_stop][['line', 'end_stop']].drop_duplicates()
        return self.conn_graph[self.conn_graph['start_stop'] == start_stop]['end_stop'].unique()

    def get_conns(self, dep_time: int, start_stop: str, end_stop: str, lines: List[str]=None):
        '''get all connections sorted by time departure'''
        conn = self.conn_graph[(self.conn_graph['start_stop'] == start_stop) & (self.conn_graph['end_stop'] == end_stop)]
        if lines:
            conn = conn[conn['line'].isin(lines)]
        time_diff = diff(conn['departure_sec'], dep_time)
        return conn.loc[time_diff.sort_values().index]

    def get_earliest_conn(self, dep_time: int, start_stop: str, end_stop: str, line: Union[str, None]=None) -> pd.Series:
        '''Returns the earliest connection between two stops, optionally by a given line'''
        conn = self.get_conns(dep_time, start_stop, end_stop, [line] if line else None)
        return conn.iloc[0]
    
    def get_all_earliest_from(self, dep_time: int, start_stop: str):
        '''Returns all earliest connections to all neighbouring stops'''
        return [self.get_earliest_conn(dep_time, start_stop, end_stop) for end_stop in self.get_neighbours(start_stop)]

    def get_all_earliest_conns(self, dep_time: int, start_stop: str, end_stop: str) -> List[pd.Series]:
        return [self.get_earliest_conn(dep_time, start_stop, end_stop, line) for line in self.get_lines(start_stop, end_stop)]

    def conn_cost(self, line: str, dep_time: int, start_stop: str, end_stop: str):
        '''Cost is the time of waiting for the transport and commuting'''
        connection = self.get_earliest_conn(dep_time, start_stop, end_stop, line)
        return (connection, connection['arrival_sec'] - dep_time)
    
    def cost(self, dep_time: int, conn: pd.Series) -> int:
        return conn['arrival_sec'] - dep_time
```

```python
graph = Graph(connection_graph)
frontier = PriorityQueue()
```

```python
def print_info(conn):
    print(f"{conn['line']} goes from {conn['start_stop']} to {conn['end_stop']} and leaves at {conn['departure_time']}, arrives at {conn['arrival_time']}")
```

```python
def sec_to_time(seconds: int) -> str:
    hour = (seconds // 3600)
    minutes = (seconds % 3600) // 60 
    secs = (seconds % 3600) % 60 
    return f'{hour:02d}:{minutes:02d}:{secs:02d}'
```

```python
start = 'KLECINA'
goal = 'Przyjaźni'
leave_hour = '20:00:00'

cost_so_far = {}
# if commuting A -> B, then this will be came_from[B] = A so we can recreate the path
came_from = {}
cost_so_far[start] = 0
came_from[start] = None

frontier.put((cost_so_far[start], (start, None)))

leave_sec = time_to_normalized_sec(leave_hour)

i = 0
while not frontier.empty():
    # get the stop with the lowest cost 
    cost, (current, line) = frontier.get()

    if current == goal:
        break 

    print(f'[{i}]')
    for next_conn in graph.get_all_earliest_from(leave_sec + cost, current):
        print_info(next_conn)
        next = next_conn['end_stop']
        # cost of commuting start --> current and current --> next 
        new_cost = cost + graph.cost(leave_sec + cost, next_conn)
        if next not in cost_so_far or new_cost < cost_so_far[next]:
            print(f'We arrive from {start} to {next} at {sec_to_time(leave_sec + new_cost)}')
            cost_so_far[next] = new_cost
            frontier.put((new_cost, (next, next_conn['line'])))
            came_from[next] = current
    i+=1
```

```python
def path_to_list(node: str, connections: dict):
    path = [node]
    while node is not None:
        node = connections[node]
        path.append(node)
    path.reverse()
    return path[1:]

path_to_list(goal, came_from)
```
